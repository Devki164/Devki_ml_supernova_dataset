```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import roc_curve,auc,confusion_matrix, classification_report
from sklearn.pipeline import Pipeline
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import label_binarize
import random
np.random.seed(42)
random.seed(42)
```

```python
class classifier:
  def RandomState(self,random_state=0):
    self.random_state = random_state
    self.thresholds={}
  def data_reforming(self,raw_data):
    #adding features using physics and data from provided columns
    self.data = raw_data.copy()
    self.data['temp_radius_ratio_log'] = np.log((self.data['st_teff']+1e-9)/(self.data['koi_prad']+1e-9))
    self.data['area_ratio_sq'] = (self.data['koi_ror']**2)*1000
    self.data['log_snr'] = np.log(self.data['koi_model_snr'])

  def threshold(self):
    self.thresholds['radius_limit'] = self.data_train['koi_prad'].quantile(0.99)
    self.thresholds['mass_limit'] = self.data_train['st_mass'].quantile(0.95)
  #adding columns which check some features that can clearly distinguish false positive from confirmed
  def threshold_implement(self,df):
    self.df = df.copy()
    self.df['radius_value'] = (self.df['koi_prad']>self.thresholds['radius_limit']).astype(int)
    self.df['impact_value'] = (self.df['koi_impact']>0.9).astype(int)
    self.df['mass_value'] = (self.df['st_mass']>self.thresholds['mass_limit']).astype(int)
    return self.df

  def data_splitting(self):
    #converting target column data to float values
    mapping = {'CANDIDATE':1.0,'FALSE POSITIVE':0.0,'CONFIRMED':2.0}
    self.data['koi_disposition'] = self.data['koi_disposition'].map(mapping)
    self.data = self.data.replace([np.inf, -np.inf], np.nan)

    self.data_train,self.data_test = train_test_split(self.data,test_size=0.3,random_state=self.random_state,stratify=self.data['koi_disposition'])
    self.threshold()
    self.data_train = self.threshold_implement(self.data_train)
    self.data_test = self.threshold_implement(self.data_test)
    self.y_train = self.data_train['koi_disposition']
    self.y_test = self.data_test['koi_disposition']
    drop_cols = ['koi_disposition','kepid']

    self.x_train = self.data_train.drop(drop_cols,axis=1)
    self.x_test = self.data_test.drop(drop_cols,axis=1)


  def model1(self):
    self.pipeline1 = Pipeline([
        ('imputer',SimpleImputer(strategy='median')),
        ('scaler',StandardScaler()),
        ('classifier',HistGradientBoostingClassifier(max_iter=200,categorical_features=None,max_leaf_nodes=50,min_samples_leaf=4,learning_rate=0.1,max_depth=5,random_state=self.random_state,class_weight='balanced'))
    ])
    return self.pipeline1

  def classification(self):
    train_a = (self.y_train==0.0)|(self.y_train==2.0)
    test_a = (self.y_test==0.0)|(self.y_test==2.0)
    self.model1a = self.model1()
    self.model1a.fit(self.x_train[train_a],self.y_train[train_a])


    train_b = (self.y_train==0.0)|(self.y_train==1.0)
    test_b = (self.y_test==0.0)|(self.y_test==1.0)
    self.model1b = self.model1()
    self.model1b.fit(self.x_train[train_b],self.y_train[train_b])


    train_c = (self.y_train==1.0)|(self.y_train==2.0)
    test_c = (self.y_test==1.0)|(self.y_test==2.0)
    self.model1c = self.model1()
    self.model1c.fit(self.x_train[train_c],self.y_train[train_c])

    self.y_pred_raw = self.model1a.predict(self.x_test)
    self.y_pred = self.y_pred_raw.copy()

    mask_fp = (self.y_pred_raw==0.0)
    if np.any(mask_fp):
      self.y_pred[mask_fp] = self.model1b.predict(self.x_test[mask_fp])
    mask_conf = (self.y_pred_raw==2.0)
    if np.any(mask_conf):
      self.y_pred[mask_conf] = self.model1c.predict(self.x_test[mask_conf])


    return self.y_pred
  def get_probas(self, X):
    prob_a = self.model1a.predict_proba(X)
    prob_b = self.model1b.predict_proba(X)
    prob_c = self.model1c.predict_proba(X)

    final_probas = np.zeros((X.shape[0], 3))
    final_probas[:, 0] = (prob_a[:, 0] + prob_b[:, 0]) / 2

    final_probas[:, 1] = (prob_b[:, 1] + prob_c[:, 0]) / 2

    final_probas[:, 2] = (prob_a[:, 1] + prob_c[:, 1]) / 2

    row_sums = final_probas.sum(axis=1)
    return final_probas / row_sums[:, np.newaxis]

  def plot_multiclass_roc(self):

    y_score = self.get_probas(self.x_test)
    y_test_bin = label_binarize(self.y_test, classes=[0.0,1.0,2.0])
    n_classes = y_test_bin.shape[1]

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    target_names = ['False Positive', 'Candidate', 'Confirmed']
    colors = ['#e74c3c', '#3498db', '#2ecc71']
    plt.figure(figsize=(10, 7))

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

        plt.plot(fpr[i], tpr[i], color=colors[i], lw=2.5,
                 label=f'ROC: {target_names[i]} (AUC = {roc_auc[i]:.3f})')


    fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel(), y_score.ravel())
    roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

    plt.plot(fpr["micro"], tpr["micro"],
             label=f'Overall Micro-average (AUC = {roc_auc["micro"]:.3f})',
             color='deeppink', linestyle=':', linewidth=4)

    plt.plot([0, 1], [0, 1], 'k--', lw=1.5)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - Specificity)')
    plt.ylabel('True Positive Rate (Sensitivity)')
    plt.title('Receiver Operating Characteristic (ROC) - Exoplanet Tournament')
    plt.legend(loc="lower right", fontsize='small')
    plt.grid(alpha=0.3)
    plt.show()

    print(f"Global Model AUC Score: {roc_auc['micro']:.4f}")

  def plot_confusion_matrix(self):
    self.cm = confusion_matrix(self.y_test,self.y_pred,labels=[2.0,1.0,0.0])
    plt.figure(figsize=(8,6))
    sns.heatmap(self.cm,annot=True,fmt='d',cmap='Blues',
                xticklabels=['confirmed','candidate','false positive'],
                yticklabels=['confirmed','candidate','false positive'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

  def classification_report(self):
    labels=[0.0,1.0,2.0]
    target = ['False Positive','Candidate','Confirmed']
    report = classification_report(self.y_test,self.y_pred,target_names=target,labels=labels)
    print('Classification report:\n')
    print(report)



```

```python
raw_data = pd.read_csv('/content/supernova_dataset.csv')
raw_data.head()
model = classifier()
model.RandomState(random_state=42)
model.data_reforming(raw_data)
model.data_splitting()
model.classification()
model.plot_multiclass_roc()

model.plot_confusion_matrix()
model.classification_report()
```

```python
def predict_new_data(self,x_new):
  self.data_reforming(x_new)
  self.threshold()
  self.x_processed = self.threshold_implement(self.data)
  drop_cols = ['kepid']
  self.x_input = self.x_processed(columns=[c for c in drop_cols if c in self.x_processed.columns],errors='ignore' )
  y_pred_raw = self.model1a.predict(self.x_input)
  y_pred = y_pred_raw.copy()
  mask_fp = (y_pred_raw==0.0)
  if np.any(mask_fp):
    y_pred[mask_fp] = self.model1b.predict(self.x_input['mask_fp'])
  mask_conf = (y_pred_raw==2.0)
  if np.any(mask_conf):
    y_pred[mask_conf] = self.model1c.predict(self.x_input['mask_conf'])
  return y_pred
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,mean_absolute_percentage_error,root_mean_squared_error
from sklearn.model_selection import train_test_split
class regressor:
  def RandomState2(self,random_state=0):
    self.random_state = random_state
    self.thresholds2={}

  def threshold2(self):

    self.thresholds2['mass_limit'] = self.data2_train['st_mass'].quantile(0.95)
  #adding columns which check some features that can clearly distinguish false positive from confirmed
  def threshold_implement2(self,df2):
    self.df2 = df2.copy()

    self.df2['impact_value'] = (self.df2['koi_impact']>0.9).astype(int)
    self.df2['mass_value'] = (self.df2['st_mass']>self.thresholds2['mass_limit']).astype(int)
    return self.df2

  def data_reforming2(self,raw_data):
    self.data2 = raw_data.copy()
    self.data2['area_ratio_sq'] = (self.data2['koi_ror']**2)*1000
    self.data2['log_snr'] = np.log(self.data2['koi_model_snr']+1e-8)

    self.data2.dropna(subset=['koi_prad'],inplace=True)
    self.data2.replace([np.inf, -np.inf],np.nan,inplace=True)

    self.data2['koi_prad'] = self.data2['koi_prad'].clip(upper=1000)
    mapping = {'CANDIDATE':1.0,'FALSE POSITIVE':0.0,'CONFIRMED':2.0}
    self.data2['koi_disposition'] = self.data2['koi_disposition'].map(mapping)

    return self.data2

  def model_traning(self):
    self.X = self.data2.drop('koi_prad',axis=1)
    self.Y = self.data2['koi_prad']

    self.X_train,self.X_test,self.Y_train,self.Y_test = train_test_split(self.X,self.Y,test_size=0.4,random_state=self.random_state)
    self.data2_train = self.X_train.copy()
    self.data2_train['koi_prad'] = self.Y_train
    self.threshold2()
    self.X_train = self.threshold_implement2(self.X_train)
    self.X_test = self.threshold_implement2(self.X_test)

    self.pipeline2 = Pipeline([
        ('scaler',StandardScaler()),
        ('Regressor',HistGradientBoostingRegressor(max_iter=400,max_leaf_nodes=40,min_samples_leaf=5,max_depth=5,loss='absolute_error',random_state=self.random_state))
    ])

    self.pipeline2.fit(self.X_train,self.Y_train)

    self.Y_pred = self.pipeline2.predict(self.X_test)

    print('Mean Absolute Error:',mean_absolute_error(self.Y_test,self.Y_pred))
    print('Mean Squared Error:',mean_squared_error(self.Y_test,self.Y_pred))
    print('Root Mean Squared Error:',np.sqrt(mean_squared_error(self.Y_test,self.Y_pred)))
    print('R2 Score:',r2_score(self.Y_test,self.Y_pred))
    print('Mean Absolute Percentage Error:',mean_absolute_percentage_error(self.Y_test,self.Y_pred))

    return self.pipeline2

  def plot_radius(self):
    plt.figure(figsize=(10,8))
    plt.scatter(self.Y_test,self.Y_pred,color='blue',label='Predicted',alpha=0.4,edgecolor='k')

    max_value = max(max(self.Y_test),max(self.Y_pred))
    min_value = min(min(self.Y_test),min(self.Y_pred))
    plt.plot([min_value,max_value],[min_value,max_value],'r--',label='Perfect Prediction',lw=2)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Actual Radius(log scale)')
    plt.ylabel('Predicted Radius(log scale)')
    plt.title('Actual vs Predicted Radius')
    plt.legend()
    plt.grid(True,alpha=0.5,which='both',ls='--')
    plt.show()
```

```python
model2 = regressor()
model2.RandomState2(random_state=42)
model2.data_reforming2(raw_data)
model2.model_traning()
model2.plot_radius()
```
